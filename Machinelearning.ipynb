```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

```python
df0 = pd.read_csv('data2.csv')
df = df0.drop(columns = ['Result'])
df
```

```python
target1 = pd.get_dummies(df0['Result'])
target1 = target1.drop(columns = ['Lost'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df, target1, random_state = 0)
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(Y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(Y_test.shape))
```

```python
#Creat Train Tensor
```

```python
x_train = np.mat(X_train.fillna(value=-1))
x_train = torch.tensor(x_train).float()
y_train = np.mat(Y_train.fillna(value=-1))
y_train = torch.tensor(y_train).float()
```

```python
#Creat Test Tensor
```

```python
x_test = np.mat(X_test.fillna(value=-1))
x_test = torch.tensor(x_test).float()
y_test = np.mat(Y_test.fillna(value=-1))
y_test = torch.tensor(y_test).float()
```

```python
#Define Neturl Network
```

```python
class myNet(nn.Module):
  
  def __init__(self):
    ''' 
      Initialization of my Neural Network
    '''

    super(myNet, self).__init__()
    
    self.layer1 = nn.Linear(24, 16)
    self.layer2 = nn.Linear(16, 8)
    self.layer3 = nn.Linear(8, 1)
    self.output = nn.Sigmoid()

  def forward(self, x):
    '''
      Forward-Propagation
    '''
    x = nn.functional.relu(self.layer1(x))
    x = nn.functional.relu(self.layer2(x))
    x = self.layer3(x)
      
    return self.output(x)
```

```python
#Setup Neural Network with Optimizer and Loss Function
```

```python
model = myNet()
optimzer = torch.optim.SGD(model.parameters(), lr = 0.001, momentum = 0.9)
loss_func = nn.MSELoss()
```

```python
#Display Neural Network Architecture
```

```python
print(model)
```

```python
#Train Neural Network
```

```python
for epoch in range(500):
  for i in range(0, len(x_train)):
    if not(torch.any(x_train[i].isnan())):
      optimzer.zero_grad() 
      out = model(x_train[i])
      loss = loss_func(out, y_train[i])
      
      loss.backward()
      optimzer.step()

      if i % 50 is 0:
        print("Network Output: ", out)
        print("Actual Label: ", y_train[i])
        print("Current Loss: ", loss.item())
```

```python
print("Output: ", model(x_test[3]))
print("Actual: ", y_test[3])
```

```python
#Test the network
```

```python
model.eval()

avg_cost = 0.0

for i in range(1, len(x_test)):
  out = model(x_test[i])
  loss = loss_func(out, y_test[i])

  avg_cost += loss.item()

print('Test Average Loss: %f' % (float(avg_cost) / float(len(x_test))))
```

```python
accurate = 0.0
for i in range(0,len(x_test)):
    if np.round(model(x_test[i]).item())==y_test[i].item():
        accurate += 1
print('Test accuracy: %f' %(float(accurate)/float(len(x_test))))
```
